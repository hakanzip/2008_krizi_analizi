```python
df_altin["Altın_Tarih"] = pd.to_datetime(df_altin["Altın_Tarih"])
df_sp500["SP500_Tarih"] = pd.to_datetime(df_sp500["SP500_Tarih"])
start_date = max(df_altin["Altın_Tarih"].min(), df_sp500["SP500_Tarih"].min())
end_date = min(df_altin["Altın_Tarih"].max(), df_sp500["SP500_Tarih"].max())
altin_filtered = df_altin[(df_altin["Altın_Tarih"] >= start_date) & (df_altin["Altın_Tarih"] <= end_date)].copy()
sp500_filtered = df_sp500[(df_sp500["SP500_Tarih"] >= start_date) & (df_sp500["SP500_Tarih"] <= end_date)].copy()
merged = pd.merge(
    altin_filtered[["Altın_Tarih", "Altın_Kapanış"]],
    sp500_filtered[["SP500_Tarih", "SP500_Kapanış"]],
    left_on="Altın_Tarih",
    right_on="SP500_Tarih")
merged["SP500_Altın_Spread"] = merged["SP500_Kapanış"] - merged["Altın_Kapanış"]

plt.figure(figsize=(14, 6))
sns.lineplot(data=merged, x="Altın_Tarih", y="SP500_Altın_Spread", label="SP500 - Altın Spread", linewidth=2.5)
plt.title("S&P 500 ve Altın Arasındaki Fiyat Farkı", fontweight="bold", fontsize=14)
plt.xlabel("Tarih", fontweight="bold")
plt.ylabel("Fiyat Farkı (USD)", fontweight="bold")
plt.gca().xaxis.set_major_locator(mdates.YearLocator(2))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

2006’dan sonra inşaat ruhsatları ve yeni konut başlangıçlarındaki keskin düşüş, konut sektöründe yaklaşan bir krizin ilk sinyali olarak değerlendirilebilir. Bu azalma, sektörde talebin zayıfladığı, yatırımın yavaşladığı ve piyasadaki ısınmanın son bulduğunu göstermektedir. Böyle erken bir yavaşlama, 2008 krizinin temel yapı taşlarından biri olmuş olabilir. Yani bu grafik, kriz öncesi balon patlamasından önce sektörün soğuduğunu, kırılganlığın arttığını açıkça ortaya koymaktadır.

```python
sp500_close = df_sp500["SP500_Kapanış"]
Q1 = np.percentile(sp500_close, 25)
Q3 = np.percentile(sp500_close, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df_sp500[(sp500_close < lower_bound) | (sp500_close > upper_bound)]

plt.figure(figsize=(14, 6))
sns.lineplot(data=df_sp500, x="SP500_Tarih", y="SP500_Kapanış", label="S&P 500 Kapanış", linewidth=2.5)
plt.axhline(lower_bound, color='red', linestyle='--', label='Alt IQR Sınırı')
plt.axhline(upper_bound, color='red', linestyle='--', label='Üst IQR Sınırı')
plt.scatter(outliers["SP500_Tarih"], outliers["SP500_Kapanış"], color='red', label='Anomali', s=50)
plt.title("S&P 500 Kapanış Fiyatında IQR Yöntemi ile Anomali Tespiti", fontweight="bold", fontsize=14)
plt.xlabel("Tarih", fontweight="bold")
plt.ylabel("Fiyat", fontweight="bold")
plt.gca().xaxis.set_major_locator(mdates.YearLocator(2))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

Bu grafikte, özellikle 1998 ve 2000 yılları arasındaki hızlı yükselişten sonra S&P 500’ün 1500 seviyesine ulaştığı ve sonrasında Dotcom balonu patlamasıyla birlikte piyasanın sert bir şekilde düştüğü gözlemlenmektedir. 2006–2008 döneminde endeksin yeniden bu 1500 seviyesine yaklaşması, dönemin yatırımcıları için yeni bir balon riski sinyali olarak değerlendirilebilirdi. Yani, geçmişte yaşanan Dotcom balonunun zirve seviyesinin yeniden test edilmesi, piyasada kırılganlık ve yaklaşan kriz endişesi yaratmıştır.

Öte yandan, IQR yöntemiyle tespit edilen anomali noktalarının büyük kısmı 2020 sonrası döneme denk gelmiştir. Bunun nedeni, S&P 500’ün pandemi sonrası çok hızlı ve yüksek seviyelere ulaşmasıdır. Bu gelişme, 2008 krizindeki ani düşüşlerin, günümüzdeki fiyat seviyeleriyle kıyaslandığında artık istatistiksel olarak “anormal” olarak algılanmamasına neden olmuştur. Dolayısıyla, IQR yöntemi tek başına geçmiş krizleri önceden yakalamada yeterli değildir; çünkü piyasa yapısı ve fiyat seviyeleri zamanla büyük değişiklik göstermiştir.

Her modelin test verisi üzerindeki tahminleriyle gerçek değerleri karşılaştırdık. Dört modelin doğruluk oranlarını hesapladık. Doğruluk skoru modelin doğru tahmin ettiği örneklerin toplam test verisine oranıdır. Her modelin performansını sayı olarak görmek için print komutuyla ekrana yazdırdık.

```python
from sklearn.metrics import classification_report, confusion_matrix
def quick_metrics(y_true, y_pred, model_name):
    report = classification_report(y_true, y_pred, output_dict=True)
    prec = round(report['1']['precision'], 2)
    rec = round(report['1']['recall'], 2)
    f1 = round(report['1']['f1-score'], 2)
    print(f"{model_name}: Precision={prec}, Recall={rec}, F1={f1}")

quick_metrics(y_test, log_pred, "Logistic Regression")
quick_metrics(y_test, rf_pred, "Random Forest")
quick_metrics(y_test, xgb_pred, "XGBoost")
quick_metrics(y_test, svm_pred, "SVM")
```

Bu fonksiyonla her modelin sadece resesyon sınıfı (1) için en önemli üç metriğini sade şekilde hesapladık. Precision, recall ve F1 skorlarını ayrı ayrı aldık ve virgülden sonra iki basamakla yuvarladık. Sonuçları model isimleriyle birlikte tek satırda yazdırdık. Bu sayede karşılaştırma yapmak kolaylaştı ve gereksiz metinlerden kurtulmuş olduk.

Çıktıya göre Logistic Regression modeli resesyon olan durumların %83’ünü yakaladı ama %71’ini doğru tahmin etti, F1 skoru dengeli çıktı. Random Forest ve XGBoost modelleri aynı sonucu verdi, bu modeller resesyonları doğru tahmin etmekte çok başarılı görünüyor, ancak fazla iyi olmaları overfit ihtimalini akla getiriyor. SVM modeli ise hiçbir resesyonu yakalayamadı, yani bu sınıfı tamamen göz ardı etti ve sıfır performans gösterdi. Bu da veri dengesizliğinde SVM’nin zayıf kaldığını gösteriyor.
